```
def read_input(input_file):

    
    logging.info("reading file {0}...this may take a while".format(input_file))
    
    with gzip.open (input_file, 'rb') as f:
        for i, line in enumerate (f): 

            if (i%10000==0):
                logging.info ("read {0} reviews".format (i))
            # do some pre-processing and return a list of words for each review text
            yield gensim.utils.simple_preprocess (line)

# read the tokenized reviews into a list
# each review item becomes a serries of words
# so this becomes a list of lists
documents = list (read_input (data_file))
logging.info ("Done reading data file")
```

```
df = pd.read_excel('mist_dataset.xlsx', header=None)
df.head()
```

```
import nltk
nltk.download('punkt')
nltk.download('stopwords')
from nltk.tokenize import word_tokenize
train = df[0]
train = train.str.replace('[^a-zA-Z]',' ').str.lower()
stop_re = '\\b'+'\\b|\\b'.join(nltk.corpus.stopwords.words('english'))+'\\b'
train = train.str.replace(stop_re, '')
###
train = [nltk.word_tokenize(sentence) for sentence in train]
#train
```

```
import nltk
nltk.download('wordnet')
from nltk.stem.wordnet import WordNetLemmatizer

#porter = PorterStemmer() 
wordnet = WordNetLemmatizer()

preprocessed_docs = []

for doc in train:
    final_doc = []
    for word in doc:
        #final_doc.append(porter.stem(word))
        final_doc.append(wordnet.lemmatize(word))
    
    preprocessed_docs.append(final_doc)
train = preprocessed_docs
#train
```

```
documents.extend(train)
```

```
model = gensim.models.Word2Vec (documents, size=150, window=10, min_count=1, workers=10)
model.train(documents,total_examples=len(documents),epochs=10)
```

```
def compute_vec(query):
    sm = np.zeros(150)
    for tok in query.split(" "):
        sm = sm + model.wv[tok]
    return sm
```

```
import numpy as np

def cos_sim(a, b):
    dot_product = np.dot(a, b)
    norm_a = np.linalg.norm(a)
    norm_b = np.linalg.norm(b)
    return dot_product / (norm_a * norm_b)
```

```
import nltk
#nltk.download('stopwords')
nltk.download('wordnet')
from nltk.corpus import stopwords
import string
from nltk.stem.wordnet import WordNetLemmatizer
from nltk import word_tokenize
nltk.download('punkt')

def clean1(text1):
  #punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
  #no_punct = ""
  #for char in text1:
   #  if char not in punctuations:
   #      no_punct = no_punct + char
  exclist = string.punctuation
  # remove punctuations and digits from oldtext
  table_ = str.maketrans('', '', exclist)
  #newtext = oldtext.translate(table_)
  text = ' '.join(text1.translate(table_).split())
  #sents = ['i like cars', 'cats are the best']
  word_list = nltk.word_tokenize(text)
  #print(word_list)
  # Lemmatize list of words and join
  lemmatizer = WordNetLemmatizer()
  lemmatized_output = ' '.join([lemmatizer.lemmatize(w) for w in word_list])
  #print(lemmatized_output)
  text=lemmatized_output
  #text = no_punct
  text = re.sub(r'\[[0-9]*\]',' ',text)
  text = re.sub(r'\s+',' ',text)
  text = text.lower()
  text = re.sub(r'\d',' ',text)
  text = re.sub(r'\s+',' ',text)
  stopwords = ['i', 'you', 'your', 'an', 'a', 'at','any', 'is','the','of','in', 'about', 'that','are','have to', 'need to','what','who','there','any', 'do', 'does']
  querywords = text.split()

  text  = [word for word in querywords if word not in stopwords]
  text = ' '.join(text)
  #stop_words = set(stopwords.words('english'))
  #no_stop = ""
  #for char in text:
   #  if char not in stop_words:
    #     no_stop = no_stop + char
  #text = no_stop  
  #print(text)
  return text
```
